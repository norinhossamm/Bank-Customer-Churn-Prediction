```python
from joblib import load
import pandas as pd
```

```python
# Load the model
model_XGB1 = load('model_XGB1.joblib')
```

```python
df = pd.read_csv("test.csv")
df
```

```python
df.shape #Checking rows and columns
```

```python
#Check for missing values:
df.isnull().sum() 

#No missing values 
```

```python
#Check for duplicates:
df.duplicated().sum()

#No duplicates
```

```python
df.info() #Type of data and count 
```

```python
#Drop not needed features
df= df.drop(['CustomerId','Surname'], axis=1)
df
```

```python
#Convert F/M to 0/1 
df['Gender']= df['Gender'].map({'Male':1, 'Female':0})
df
```

```python
#One hot encoding for countries in geography 
df = pd.get_dummies(df, columns=['Geography'], dtype=int)
df
```

```python
df['ZeroBalance'] = df['Balance'].apply(lambda x: 1 if x == 0 else 0)
df['IsSenior'] = df['Age'].apply(lambda x: 1 if x >= 60 else 0)
df['IsActive_by_CreditCard'] = df['HasCrCard'] * df['IsActiveMember']
df['ActiveMem_noProducts'] = df['NumOfProducts'] * df['IsActiveMember']
df['Credit_per_tenure'] = df['Tenure'] / df['CreditScore']
df['Salary_per_tenure'] = df['Tenure'] / df['EstimatedSalary']
df['Salary_Age_ratio'] = df['EstimatedSalary'] / df['Age']
df['Balance_Age_Ratio'] = df['Balance'] / df['Age']
df['Balance_Salary_Ratio'] = df['Balance'] / df['EstimatedSalary']
df['Products_Per_Tenure'] =  df['Tenure'] / df['NumOfProducts']
df
```

```python
#Drop features
df= df.drop(['Tenure','IsActiveMember','Balance','EstimatedSalary'], axis=1)
df
```

```python
X = df.drop(columns='id', axis=1) #Inputs
X
```

```python
y_pred= model_XGB1.predict_proba(X)
PREDICTIONS= []
for i in range (len(y_pred)):
    PREDICTIONS.append(y_pred[i][1])
PREDICTIONS
```

```python
df['Exited']= PREDICTIONS
df
```

```python
selected_columns = df[['id', 'Exited']]

# Save the selected columns to a new CSV file
selected_columns.to_csv('Results.csv', index=False)
```
